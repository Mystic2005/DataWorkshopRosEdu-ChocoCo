```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/Chocolate_Sales_Curated.csv")
```

```python
df["date"] = pd.to_datetime(df["date"], errors='coerce')
df["month"] = df["date"].dt.month
df["day_of_week"] = df["date"].dt.dayofweek
df
```

```python
# Prepare for regression
from sklearn.preprocessing import OneHotEncoder

target = "amount"
inputs = ["boxes_shipped", "country", "product", "sales_person", "month", "day_of_week"]

X = df[inputs]
y = df[target]

categorical_cols = ["country", "product", "sales_person"]
encoder = OneHotEncoder(handle_unknown="ignore", sparse_output=False)

# Fit și transform
encoded = encoder.fit_transform(X[categorical_cols])
encoded_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(categorical_cols))

numeric_cols = ["boxes_shipped", "month", "day_of_week"]
X_encoded = pd.concat([X[numeric_cols].reset_index(drop=True), encoded_df], axis=1)
```
