```python
import pandas as pd
import numpy

df = pd.read_csv("../data/raw/Chocolate_Sales.csv")
df
```

```python
# Curating the dataframe
df.columns = df.columns.str.lower().str.replace(" ", "_")
df["date"] = pd.to_datetime(df["date"], format="%d/%m/%Y")
df["amount"] = pd.to_numeric(df["amount"].str.replace("$", "").str.replace(",", ""))
df["boxes_shipped"] = pd.to_numeric(df["boxes_shipped"])
df = df.drop_duplicates()
```

```python
# Saving the processed output
df.to_csv("../data/processed/Chocolate_Sales_Curated.csv", index=False)
```
